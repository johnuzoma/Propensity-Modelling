# Model Building

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

# Make predictive features to be all columns except "UserID" and "ordered"
features = train.drop(['UserID', 'ordered'], axis=1)

# Our target variable is "ordered"
target = train['ordered']

# Split the data into 80% train and 20% test
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    train_size=0.8,
    test_size=0.2
)
```

### Check for class imbalance

```python
# Import useful functions from matplotlib.ticker
from matplotlib.ticker import FormatStrFormatter, StrMethodFormatter, PercentFormatter

# Initialize a matplotlib figure with set dimensions
fig, ax = plt.subplots(figsize =(8, 6))

# Create dictionary that maps binary values in "ordered" to descriptive names
data = {0: 'No order', 1: 'Ordered'}

# Get the value counts for each unique value in y_train
value_counts = y_train.replace(data).value_counts()

# Calculate the total count
total_count = len(y_train)

# Calculate the percentage of each value
percentages = value_counts / total_count * 100

# Create the histogram with two bars
plt.bar(range(len(percentages)), percentages, align='center', color=['red', 'green'])

# Set the x-axis ticks and labels
plt.xticks(range(len(value_counts)), value_counts.index)

# Set the y-axis formatter to display percentages
plt.gca().yaxis.set_major_formatter(PercentFormatter())

# Add data labels to the bars
for i, count in enumerate(percentages):
    plt.text(i, count, f'{percentages[i]:.0f}%', ha='center', va='bottom')

# Add title
plt.title('Relative Distribution Of Order Status In Train Set')

plt.show()
```

Looks like we have more visitors who didn't order (96%) than visitiors who ordered (4%), signifying a class imbalance which is not good for an ML model. This is not good because the imbalance can make the proposed ML model to always predict a specific class with the highest occurrence, in this case, "No order".

Predicting a no-order when the actual is order, will lead to loss of potential revenue, and allocating marketing resources based on inaccurate predictions can lead to inefficient use of the firm's budget. To resolve this class imbalance, I will undersample the data.

### Undersampling training data

```python
# Concatenate X_train and y_train
X_y_train = pd.concat([
    X_train.reset_index(drop = True),
    y_train.reset_index(drop = True)
], axis = 1)

# Assign counts of no-orders and orders
count_non_orders, count_orders = X_y_train['ordered'].value_counts()
```

```python
# Create data sets for orders and no-orders
orders = X_y_train[X_y_train['ordered'] == 1]
non_orders = X_y_train[X_y_train['ordered'] == 0]

# Undersample the no-orders since it's the greater number
non_orders_under = non_orders.sample(count_orders)

# Concatenate the undersampled no-orders with orders
X_y_train_under = pd.concat([
    non_orders_under.reset_index(drop = True),
    orders.reset_index(drop = True)
], axis = 0)
```

```python
# Initialize a matplotlib figure with set dimensions
fig, ax = plt.subplots(figsize =(8, 6))

# Create dictionary that maps binary values in "ordered" to descriptive names
data = {0: 'No order', 1: 'Ordered'}

# Get the value counts for each unique value in y_train
value_counts = X_y_train_under['ordered'].replace(data).value_counts()

# Calculate the total count
total_count = len(X_y_train_under['ordered'])

# Calculate the percentage of each value
percentages = value_counts / total_count * 100

# Create the histogram with two bars
plt.bar(range(len(percentages)), percentages, align='center', color=['red', 'green'])

# Set the x-axis ticks and labels
plt.xticks(range(len(value_counts)), value_counts.index)

# Set the y-axis formatter to display percentages
plt.gca().yaxis.set_major_formatter(PercentFormatter())

# Add data labels to the bars
for i, count in enumerate(percentages):
    plt.text(i, count, f'{percentages[i]:.0f}%', ha='center', va='bottom')

# Add title
plt.title('Relative Distribution Of Order Status In Train Set')

plt.show()
```

Now the training set has the same number of orders and no-orders.

Note that I didn't undersample the test set because the test set represents the type of data that will be seen by the model in the real world, so changing it would test the model on unrealistic data.

```python
# Now splitting undersampled training data
features = X_y_train_under.drop('ordered', axis=1)
target = X_y_train_under['ordered']
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    train_size=0.8,
    test_size=0.2
)
```

```python
# Get the count of each order status after 2nd split
y_train.value_counts()
```

This is much better as the difference between both counts is less than 10.

### Using best model recommended by TPOT

Since LinearSVC, which was recommended by TPOT, does not have the functionality to predict probabilities, I will use an SVC model with linear kernel instead. They are almost the same. In addition to whole class prediction, the latter can also predict probabilities when we set "probability=True" as an argument. We will predict probabilities later when we want to predict on previous day's visitors.

```python
# Import libraries
from sklearn.svm import SVC
from sklearn.calibration import CalibratedClassifierCV

# Create and train the SVC model with linear kernel
svc_model = SVC(C=1.0, kernel='linear', probability=True, tol=0.001)
calibrated_svc = CalibratedClassifierCV(estimator=svc_model, method='sigmoid')
calibrated_svc.fit(X_train, y_train)

# Predict probabilities
predictions = calibrated_svc.predict(X_test)
```

```python
# Import accuracy score function
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, predictions))
```

About 99% of order statuses were predicted correctly.

### Feature importance

The SVC model with a linear kernel, does not inherently provide feature importances like some other models (e.g., tree-based models).

However, I can use techniques like permutation importance to approximate feature importance for this model. Permutation importance works by shuffling the values of a single feature and measuring the impact on the model's performance. Features that, when shuffled, significantly degrade the model's performance are considered more important.

```python
# Import permutation importance function from sklearn
from sklearn.inspection import permutation_importance

# Calculate permutation importances based on accuracy score
PI = permutation_importance(calibrated_svc, X_test, y_test, scoring='accuracy', n_repeats=30, random_state=42)

# Get feature importances and their indices
importances = PI.importances_mean
feature_names = X_test.columns  # Assuming your features have names
indices = np.arange(len(importances))

# Sort features by importance in ascending order
sorted_indices = np.argsort(importances)

# Plot the horizontal bar chart with decreasing permutation importance
plt.barh(indices, importances[sorted_indices])
plt.yticks(indices, feature_names[sorted_indices])
plt.xlabel('Permutation Importance')
plt.ylabel('Feature Name')
plt.title('Permutation Importance of Features')
plt.show()
```
